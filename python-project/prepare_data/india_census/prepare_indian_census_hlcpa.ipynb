```python
import numpy as np
import pandas as pd

df = pd.read_csv("./india_census_housing-hlpca-full.csv")

print(df.shape)
df.head()
```

```python
missing_data = df.isnull()

true_counts = [(column, missing_data[column].values.sum()) for column in missing_data.columns]
false_counts = [(column, (~missing_data[column].values).sum()) for column in missing_data.columns]
true_counts.sort(key=lambda x:x[1], reverse = True)
true_counts
```

```python
unique_counts = [(column, df[column].unique().size) for column in df.columns]
unique_counts.sort(key=lambda x:x[1], reverse = True)
unique_counts
```

```python
# Remove categorical variables with too many values
df.drop(["TC_VC", "Total", "Total Number of households", "Ward No", "Town Code/Village code", "Tehsil Code", 
         "State Code", "State Name", "District Code", "District Name" ,"Tehsil Name", "Area Name"], inplace=True, axis=1)
df.shape
```

```python
df.to_csv("census_india.csv", index=False)
```
