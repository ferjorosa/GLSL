```python
from scipy.io import arff
import pandas as pd
import numpy as np

data = arff.loadarff("qsar_biodeg.arff")
data = pd.DataFrame(data[0])
str_data = data.select_dtypes([np.object])
str_data = str_data.stack().str.decode('utf-8').unstack()
data["class"] = str_data["class"]

data.head()
```

```python
unique_counts = [(column, data[column].unique().size) for column in data.columns]
unique_counts.sort(key=lambda x:x[1], reverse = True)
unique_counts
```

```python
data.to_csv("new_qsar_biodeg.csv", index=False)
```
